```python
import numpy as np

def fakeSpectrum(x) :
    return (x-349.9)/215.0 * np.exp(-0.225*(x-349.9)/215.0) * np.log((x-349.9)/215.0) + 1.0

def fakeTrough(x,mu,sigma) :
    return 0.2*fakeSpectrum(mu)*np.exp(-(x-mu)**2/2.0/sigma)

def generateFalseData(NPoints, MSpectra, KEndmembers, noiseV=0, noiseB=0) :
    
    # create false abundances (breakup 1 into K parts)
    A = np.random.random([KEndmembers, NPoints])
    sumA = np.sum(A,axis=0)
    A = A/sumA

    # create false pure spectra
    XF = np.linspace(350.0,2500.0,num=MSpectra)
    print(fakeSpectrum(XF))
    F = np.zeros([KEndmembers,MSpectra]) + fakeSpectrum(XF)
    centers = (np.random.random([KEndmembers])*8.5 + 1.0)*215.0 + 349.9
    F = F - fakeTrough((np.zeros([KEndmembers,MSpectra])+XF).T,centers,1000).T
    
    # add random noise to I
    noise = noiseV*np.random.randn(NPoints, MSpectra) + noiseB
    I = np.dot(A.T,F) + noise
    
    # normalize spectra for I and F
    normsI = np.sum(I,axis=1)
    I = (I.T/normsI).T
    normsF = np.sum(F,axis=1)
    F = (F.T/normsF).T
    
    return XF,A,F,I,centers

def generatePureFs(MSpectra, centers, noiseV=0, noiseB=0) :
    
    # create false pure spectra
    XF = np.linspace(350.0,2500.0,MSpectra)
    F = np.zeros([len(centers),MSpectra]) + fakeSpectrum(XF)
    F = F - fakeTrough((np.zeros([len(centers),MSpectra])+XF).T,np.array(centers),0.1).T
    
    # add random noise
    noise = noiseV*np.random.randn(len(centers), MSpectra) + noiseB
    
    # normalize spectrum
    F = F + noise
    normsF = np.sum(F,axis=1)
    F = (F.T/normsF).T
    
    return F
```
